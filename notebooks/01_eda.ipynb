Explaratory Data Analysis (EDA)

1.Imports

```python
import pandas as pd
import numpy as np
```

2.Load Dataset

```python
df = pd.read_csv("../data/raw/isizulu_authors_dataset_cleaned.csv")
```

3.Dataset Shape

```python
print("=== DATASET SHAPE ===")
print(df.shape)
```

4.Columns

```python
print("=== COLUMNS ===")
print(df.columns.tolist())
```

5.Author Distribution

```python
print("=== AUTHOR DISTRIBUTION ===")

vc = df["author"].value_counts()

print(vc)
```

6.Character Length Statistics

```python
df["char_len"] = df["text"].str.len()

print("=== TEXT LENGTH STATS (chars) ===")
print(df["char_len"].describe())
```

7.Per Author Statistics

```python
print("=== PER AUTHOR LENGTH STATS ===")

print(
    df.groupby("author")["char_len"]
    .agg(["mean", "min", "max", "count"])
    .round(0)
)
```

8.Sample Texts

```python
print("=== SAMPLE TEXTS ===")

for author in df["author"].unique():
    sample = df[df["author"] == author]["text"].iloc[0][:200]

    print(f"\n--- {author} ---")
    print(sample)
```

9.Missing Values

```python
print("=== MISSING VALUES ===")
print(df.isnull().sum())
```

10:Duplicate Detection

```python
print("=== DUPLICATE TEXTS ===")

duplicates = df.duplicated(subset=["text"]).sum()

print("Exact duplicates:", duplicates)
```
